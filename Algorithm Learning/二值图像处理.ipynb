```python
"""
  Created on Wed Dec  4 19:09:55 2021
  二值图像处理
  @author: 余
"""
import matplotlib.pyplot as plt    #导入pyplot模块
img=plt.imread(r'..\image\1.jpg') #读取指定路径下的PNG图片(相对路径../)
plt.subplot(1,2,1)
plt.imshow(img)                                        #显示图片
#print(img)                                             #打印img数组
print(type(img))                                       #输出是数组类型
print(img.shape)                                       #输出图片大小(像素/分辨率：长*宽*3维数组)
#print(img[2000:2500,500:1000,:])                       #观察图片白色部分的数值
hight,width,length=img.shape                           #获取图片的高、宽，像素点值
img1=img.copy()                                        #复制新的数组对象
for h in range(hight):
    for w in range(width):
        for l in range(length):
            if img1[h,w,l]>=100:                       #一般折中255/2=127
                img1[h,w,l]=255                        #白色
            else:
                img1[h,w,l]=0                           #黑色
plt.subplot(1,2,2)
plt.imshow(img1)
```

```python
#matplotlib.pyplot基本操作（对照matlab）
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0.0,2.0,0.1)
s=np.sin(t*np.pi)
plt.subplot(2,2,1) #要生成两行两列，这是第一个图plt.subplot('行','列','编号')
plt.plot(t,s,'b--')
plt.ylabel('y1')
plt.subplot(2,2,2) #两行两列,这是第二个图
plt.plot(2*t,s,'r--')
plt.ylabel('y2')
plt.subplot(2,2,3)#两行两列,这是第三个图
plt.plot(3*t,s,'m--')
plt.subplot(2,2,4)#两行两列,这是第四个图
plt.plot(4*t,s,'k--')
plt.show()
```
