```python
"""
  Created on Wed Dec  4 19:09:55 2021
  
  @author: 
"""
import matplotlib.pyplot as plt    #pyplot
img=plt.imread(r'..\image\1.jpg') #PNG(../)
plt.subplot(1,2,1)
plt.imshow(img)                                        #
#print(img)                                             #img
print(type(img))                                       #
print(img.shape)                                       #(/**3)
#print(img[2000:2500,500:1000,:])                       #
hight,width,length=img.shape                           #
img1=img.copy()                                        #
for h in range(hight):
    for w in range(width):
        for l in range(length):
            if img1[h,w,l]>=100:                       #255/2=127
                img1[h,w,l]=255                        #
            else:
                img1[h,w,l]=0                           #
plt.subplot(1,2,2)
plt.imshow(img1)
```
